```python
pd.set_option("display.max_columns", None)

df = pd.read_csv("data/train.csv")

df.head(20)
df.tail(20)
```

```python
def preprocessing(data):
    cat_cols = data.select_dtypes(include=["object", "category", "str"]).columns
    data[cat_cols] = data[cat_cols].fillna("None").astype(str)

    num_cols = data.select_dtypes(include=["int64", "float64"]).columns
    data[num_cols] = data[num_cols].fillna(data[num_cols].median())

    data["log_GrLivArea"] = np.log1p(data["GrLivArea"])
    data["IsHugeHouse"] = (data["GrLivArea"] >= data["GrLivArea"].quantile(0.95)).astype(int)

    data["Age"] = 2024 - data["YearBuilt"]
    data["RemodAge"] = 2024 - data["YearRemodAdd"]

    data["IsLuxury"] = (
        (data["OverallQual"] >= 9)
        & (data["log_GrLivArea"] > data["log_GrLivArea"].quantile(0.75))
    ).astype(int)

    drop_cols = ["Id", "MiscVal", "Fence", "Alley", "GrLivArea"]
    data = data.drop(columns=[c for c in drop_cols if c in data.columns])

    return data
```

```python
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import train_test_split

X = df.copy()
X = preprocessing(X)
y = np.log1p(X.pop("SalePrice"))

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)

cat_features = X_train.select_dtypes(include=["object", "category", "str"]).columns.tolist()

X_train.head()
```

```python
model = CatBoostRegressor(
    random_state=0,
    learning_rate=0.05,
    iterations=2000,
    loss_function="RMSE",
    eval_metric="MAE",
    verbose=200
)

model.fit(
    X_train, y_train,
    cat_features=cat_features,
    eval_set=(X_valid, y_valid),
    early_stopping_rounds=50
)
```

```python
import matplotlib.pyplot as plt

feat_imp = model.get_feature_importance(prettified=True)

plt.figure(figsize=(10, 6))
plt.barh(feat_imp['Feature Id'][:25], feat_imp['Importances'][:25])
plt.gca().invert_yaxis()
plt.show()
```

```python
preds = np.expm1(model.predict(X_valid))
y_valid_real = np.expm1(y_valid)
errors = preds - y_valid_real
print(f"Mean error {errors.mean()}")
print(errors)
```

```python
test_df = pd.read_csv("data/test.csv")
test_ids = test_df["Id"]
test_df = preprocessing(test_df)

test_preds = np.expm1(model.predict(test_df))


submission = pd.DataFrame({
    "Id": test_ids,
    "SalePrice": test_preds
})

submission.to_csv("submission.csv", index=False)
```
